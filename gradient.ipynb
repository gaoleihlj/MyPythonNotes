```python
%matplotlib inline
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y = np.meshgrid(np.arange(-2,2,.2), np.arange(-2,2,.2))
z = x*np.exp(-x**2-y**2)
z.shape
```

```python
v,u = np.gradient(z,.2,.2)
fig, ax = plt.subplots()
ax.contour(x,y,z)
q = ax.quiver(x,y,u,v)
plt.show()
```

```python
x = torch.tensor(x,requires_grad=True)
y = torch.tensor(y,requires_grad=True)
```

```python
z = x*torch.exp(-x**2-y**2)
```

```python
z.backward(torch.ones_like(x))
```

```python
plt.contour(x.detach().numpy(),y.detach().numpy(),z.detach().numpy())
```

```python
print(x.grad)
print(y.grad)
```

```python
fig, ax = plt.subplots()
ax.contour(x.detach().numpy(),y.detach().numpy(),z.detach().numpy())
q = ax.quiver(x.detach().numpy(),y.detach().numpy(),x.grad.numpy(),y.grad.numpy())
plt.show()
```
